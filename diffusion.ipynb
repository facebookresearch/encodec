```python
import soundfile as sf
import torch
import IPython.display as ipd
import encodec

MBD = encodec.MultiBandDiffusion.get_mbd_24khz()
wav, sr = sf.read('/private/home/robinsr/sample_24k_robin/agitations.wav')
wav_torch = torch.from_numpy(wav).view(1, 1, -1).float().cuda()

# Regenerate 
regenerated_diffusion = MBD.regenerate(wav_torch)
ipd.display(ipd.Audio(regenerated_diffusion.squeeze(0).cpu(), rate=sr))

# Generate embs 
condition = MBD.get_condition(wav_torch)
generated = MBD.generate(condition)
ipd.display(ipd.Audio(generated.squeeze(0).cpu(), rate=sr))
```
